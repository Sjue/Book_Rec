## Popular Authors based on Book Ratings

Top ten book authors with the highest sum of Ratings

```python
pop_authors= sampled.groupby(['Book-Author'])['Book-Rating'].sum().sort_values(ascending=False).head(10)
pop_authors
```

```python
pop_authors_names = sampled[sampled['Book-Author'].isin(pop_authors.index)]
pop_authors_names
```

```python
author_rating=pop_authors_names.groupby(['Book-Author'])['Book-Rating'].mean().sort_values(ascending=False)
author_rating
```

```python
sns.barplot(x = author_rating.index, y= author_rating.values)
plt.xticks(rotation = 90)
```

#### Based on the highest Rated authors, Which book is their most 'popular'?

```python
author_title_rating = pop_authors_names.groupby(['Book-Author','Book-Title'])['Book-Rating'].sum().reset_index()
```

```python
top_books_by_author = author_title_rating.sort_values(['Book-Author', 'Book-Rating'], ascending=[True, False])
```

```python
top_books_by_author = top_books_by_author.drop_duplicates(subset = 'Book-Author', keep= 'first')
```

```python
top_books_by_author
```

```python
top_book_by_author_rating = sampled[sampled['Book-Title'].isin(top_books_by_author['Book-Title'])]
#top_book_by_author_rating
```

```python
book_ratings = top_book_by_author_rating.groupby('Book-Title')['Book-Rating'].mean().sort_values(ascending=False)
book_ratings
```

```python
sns.barplot(x= book_ratings.index, y=book_ratings.values)
plt.xticks(rotation = 90)
plt.title("Popular Author's most popular book")
```

## Which Books are considered popular by Rating 

Getting Sum by ratings to see which book has the highest amount of positive ratings. Top Ten Books with the highest ratings

```python
ratings_sum = sampled.groupby(['Book-Title'])['Book-Rating'].sum().sort_values(ascending=False)
popular_books = ratings_sum.head(10)
popular_books
```

Getting a count of top ten books that had the most ratings

```python
ratings_count=sampled.groupby(['Book-Title'])['Book-Rating'].count().sort_values(ascending=False).head(10)  
ratings_count
```

```python
sns.barplot(x = popular_books.index, y = popular_books.values)
plt.xticks(rotation = 90)
plt.title('Sum of the highest rated books')
```

```python
sns.barplot(x = ratings_count.index, y = ratings_count.values)
plt.xticks(rotation = 90)
plt.title('Count of the most rated books')
```

When Comparing the Count vs SUM, only three books differ between th two: The Painted House, The Nanny Diaries: A Novel, and the Summons. We also see that while Wild Animus is considered the lowest for the top ten Highest SUM ratings it has the second most ratings in general. The lovely bones is first in Both SUM and COUNT ratings.
